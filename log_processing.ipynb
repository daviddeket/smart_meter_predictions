```python
# Clean up logs
def remove_backspace_and_lines(file_path):
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()

        modified_lines = []
        for line in lines:
            if '\x08' not in line and '2764' not in line:
                modified_lines.append(line)

        new_file_path = file_path.replace('.log', '_modified.log')

        with open(new_file_path, 'w') as new_file:
            new_file.writelines(modified_lines)

        print(f"Modified text saved to: {new_file_path}")
    except FileNotFoundError:
        print("File not found. Please provide a valid file path.")


# Example usage
file_path = "logs/2'broadband router', 'laptop computer', 'external hard disk', 'computer'.log"  # Replace with your input file path

remove_backspace_and_lines(file_path)
```
