```python
import seaborn as sns
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

# Loading the Iris dataset
X, y = load_iris()["data"], load_iris()["target"]
# Concatenating the data in a dataFrame
df = pd.DataFrame(np.concatenate([X,y.reshape(-1,1)], axis=1),
                  columns=load_iris()["feature_names"]+["traget"])
df.head(5)
```

```python
g = sns.jointplot(x="sepal length (cm)", y="sepal width (cm)", data=df,
    kind="reg", truncate=False,
    xlim=(3.5, 9), ylim=(1, 5),
    color="m", height=5)
```

```python
#sns.jointplot?
sns.heatmap(df.corr(), annot=True)
```

```python
sns.pairplot(data=df, hue='traget', height=1.5)
```
